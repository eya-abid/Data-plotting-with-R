```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Graph

```python
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
plt.plot(x,y)
plt.show()
```

## Add Some Details

```python
plt.plot(x, y, label='linear')  # Plot some data on the axes.
plt.xlabel('x label')  # Add an x-label.
plt.ylabel('y label')  # Add a y-label.
plt.title("Simple Plot",{'fontsize': 10,'color':'red'})  # Add a title.


#another curve
x1=np.linspace(0,10)
plt.plot(x1,np.sin(x1)+8,label='sin x')

#plt.ylim(-1,1)
#plt.xlim(0,2)

plt.legend()  # Add a legend.
plt.legend(loc='lower center',ncol=2)

plt.grid()
plt.show()
```

## Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y,c=colors,s=area)
plt.colorbar()
plt.show()
```

## Bar chart

```python
labels = ['Amy', 'Rob', 'Dimitri', 'Sarah', 'Brian']
values = [60, 56, 63, 75, 48]
values2 = [30,20, 100, 90,40]
#Change the bar colors here
plt.bar(labels, values,label='v1' )#,color=['red', 'blue', 'purple', 'green', 'lavender'])
plt.bar(labels, values2,label='v2',bottom=values)

plt.ylabel('Typing Speed (wpm)')
plt.legend()
plt.show()
```

## Pie Chart

```python
labels = 'usa', 'Russia', 'china', 'tunisia'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0.3, 0) 

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90) 

plt.show()
```

## Multiple charts In One figure

```python
fig1, f1_axes = plt.subplots(ncols=2, nrows=2)
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
f1_axes[0,0].plot(x, y, label='linear')  # Plot some data on the axes.
f1_axes[0,0].set_xlabel('x label')  # Add an x-label.
f1_axes[0,0].set_ylabel('y label')  # Add a y-label.
f1_axes[0,0].set_title("Simple Plot",{'fontsize': 25,'color':'red'})  # Add a title.


#another curve
#x1=np.linspace(0,10)
f1_axes[0,0].plot(x,np.sin(x),label='sin x')

#f1_axes[0,0].ylim(-1,1)

f1_axes[0,0].legend()  # Add a legend.
f1_axes[0,0].legend(loc='lower center')

#f1_axes[0,0].grid()
fig1
```
